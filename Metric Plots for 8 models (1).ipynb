```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# create a dataframe with the model metrics
data = {'Model': ['Decision Tree', 'Random Forest', 'Support Vector Machine', 'Lightgbm', 'CatBoostRegressor', 'GradientBoostingRegressor', 'XGBRegressor', 'ANN'],
        'MSE': [0.000762, 0.000348, 0.002823, 0.000498, 0.000211, 0.000197, 0.000557, 0.020371],
        'RMSE': [0.027601, 0.018654, 0.053131, 0.022314, 0.014501, 0.014049, 0.023597, 0.142723],
        'R2': [0.903307, 0.955832, 0.641708, 0.936801, 0.973311, 0.974945, 0.929321, -1.585518]}
df = pd.DataFrame(data)

# set the Model column as index
df.set_index('Model', inplace=True)

# create a figure with 3 subplots
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# set background color of subplots
for ax in axs:
    ax.set_facecolor('#f5f5f5')
    ax.grid(color='white', linestyle='-', linewidth=1)

# plot the MSE for each model
sns.barplot(ax=axs[0], x=df.index, y=df['MSE'], palette='Set2')
axs[0].set_title('MSE')
axs[0].set_ylabel('')

# plot the RMSE for each model
sns.barplot(ax=axs[1], x=df.index, y=df['RMSE'], palette='Set2')
axs[1].set_title('RMSE')
axs[1].set_ylabel('')

# plot the R2 for each model
sns.barplot(ax=axs[2], x=df.index, y=df['R2'], palette='Set2')
axs[2].set_title('R2')
axs[2].set_ylabel('')

# add a legend to the plot
handles, labels = axs[0].get_legend_handles_labels()
fig.legend(handles, labels, loc='upper right')

# set the x-axis labels to be rotated 45 degrees for better readability
for ax in axs:
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right')

# set the plot title
fig.suptitle('Model Comparison')

# adjust the layout
plt.tight_layout()

# show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# create a dataframe with the model metrics
data = {'Model': ['Decision Tree', 'Random Forest', 'Support Vector Machine', 'Lightgbm', 'CatBoostRegressor', 'GradientBoostingRegressor', 'XGBRegressor', 'ANN'],
        'MSE': [0.000762, 0.000348, 0.002823, 0.000498, 0.000211, 0.000197, 0.000557, 0.020371],
        'RMSE': [0.027601, 0.018654, 0.053131, 0.022314, 0.014501, 0.014049, 0.023597, 0.142723],
        'R2': [0.903307, 0.955832, 0.641708, 0.936801, 0.973311, 0.974945, 0.929321, -1.585518]}
df = pd.DataFrame(data)

# set the Model column as index
df.set_index('Model', inplace=True)

# create a figure and axes
fig, ax = plt.subplots(figsize=(12, 8))

# create a color palette for each model
colors = sns.color_palette('Set2', len(df))

# plot the MSE, RMSE, and R2 for each model
df.plot(kind='barh', ax=ax, color=colors)
ax.set_title('Model Comparison')
ax.set_xlabel('Score')
ax.set_ylabel('Model')

# add the legend to the plot
handles, labels = ax.get_legend_handles_labels()
ax.legend(reversed(handles), reversed(labels), loc='lower right')

# show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Model names
models = ['Decision Tree', 'Random Forest', 'SVM', 'LightGBM', 'CatBoost', 'Gradient Boosting', 'XGBoost', 'ANN']

# Model metrics
rmse = [0.027601, 0.018654, 0.053131, 0.022314, 0.014501, 0.014049, 0.023597, 0.142723]
r2 = [0.903307, 0.955832, 0.641708, 0.936801, 0.973311, 0.974945, 0.929321, -1.585518]

# Set up the figure
fig, ax = plt.subplots(figsize=(12,8))

# Set bar width and positions
width = 0.4
x = np.arange(len(models))

# Plot RMSE and R2
rects1 = ax.bar(x - width/2, rmse, width, label='RMSE', color='steelblue')
rects2 = ax.bar(x + width/2, r2, width, label='R2', color='coral')

# Add some text for labels, title and axes ticks
ax.set_ylabel('Metric Score')
ax.set_title('Comparison of Model Performance')
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.legend()

# Add the actual values above the bars
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{:.3f}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=10)

autolabel(rects1)
autolabel(rects2)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the data
models = ['Decision Tree', 'Random Forest', 'SVM', 'Lightgbm', 'CatBoost', 'Gradient Boosting', 'XGBoost', 'ANN']
mse = [0.000762, 0.000348, 0.002823, 0.000498, 0.000211, 0.000197, 0.000557, 0.020371]
rmse = [0.027601, 0.018654, 0.053131, 0.022314, 0.014501, 0.014049, 0.023597, 0.142723]
r2 = [0.903307, 0.955832, 0.641708, 0.936801, 0.973311, 0.974945, 0.929321, -1.585518]

# Create the plot
fig, ax = plt.subplots(figsize=(12, 6))
index = np.arange(len(models))
bar_width = 0.25
opacity = 0.8

rects1 = ax.bar(index, mse, bar_width,
                alpha=opacity,
                color='b',
                label='MSE')

rects2 = ax.bar(index + bar_width, rmse, bar_width,
                alpha=opacity,
                color='g',
                label='RMSE')

rects3 = ax.bar(index + 2*bar_width, r2, bar_width,
                alpha=opacity,
                color='r',
                label='R2')

ax.set_xlabel('Models')
ax.set_ylabel('Metric Scores')
ax.set_title('Comparison of Model Performance')
ax.set_xticks(index + bar_width)
ax.set_xticklabels(models, rotation=45)
ax.legend()

plt.tight_layout()
plt.show()
```
